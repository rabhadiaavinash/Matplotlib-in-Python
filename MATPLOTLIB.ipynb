```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# creating a canvas

fig = plt.figure()
# control left,right , width and height of the canvas be (0,1)

axis = fig.add_axes([0.5,0.1,0.8,0.8])

axis.plot(x,y,'r')
```

```python
# subgraphs

fig, axes = plt.subplots(nrows =1 , ncols =2)
for ax in axes : 
    ax.plot(x,y,'r')
```

```python
# subgraphs

fig, axes = plt.subplots(nrows =2 , ncols =1)
for ax in axes : 
    ax.plot(x,y,'r')
```

```python
# figure inside figure

fig = plt.figure()
# (b/w 0 and 1  left,right.width,height)
axis1 = fig.add_axes([.1,.1,.8,.8])# big axes
axis2 = fig.add_axes([.2,.5,.4,.2]) # small canvas

axis1.plot(x,y,'r')
axis2.plot(y,x,'g')

```

```python
# dpi = dots per inch

fig =plt.figure(figsize= (16,9), dpi = 300)

fig.add_subplot()

plt.plot(x,y,'r')
```

```python
# same as the above

fig, axis = plt.subplots(figsize = (16,9), dpi =300)
axis.plot(x,y,'r')
```

```python
fig, axis = plt.subplots(figsize = (16,9), dpi =300)
axis.set_title("Avinash Plot")
axis.set_xlabel("X axis")
axis.set_ylabel("Y axis")
axis.plot(x,y,'m')
```

```python
#ax.legend(["Label1","Label2"])

fig, axis = plt.subplots(figsize = (16,9), dpi =300)
axis.set_title("Avinash Plot")
axis.set_xlabel("X axis")
axis.set_ylabel("Y axis")
#axis.plot(x,y,'m')

axis.plot(x,x**2)
axis.plot(x,x**3)
axis.legend([" plot 1","plot2"], loc =2)
```

```python
# adjust alpha value
fig, axis = plt.subplots(dpi=150)
axis.plot(x,x**2 ,color="red",alpha=0.5)
axis.plot(x,x+2 ,color="blue",alpha=0.8)
axis.plot(x,x+5 ,color="green",alpha=1 , linewidth=3, linestyle = ":" , marker = "o" , markersize = 4, markerfacecolor ="red")
```

```python
fig, Axes = plt.subplots(1,2,figsize=(8,5))

Axes[0].plot(x,x**2,x,x**3, lw=3)
Axes[0].grid(True)

Axes[1].plot(x,x**2,x,x**3, lw=3)
Axes[1].grid(True)
Axes[1].set_ylim([0,60])
Axes[1].set_xlim([2,5])

```

```python
import seaborn as sns
```

```python
d = sns.load_dataset("tips")
```

```python
d.head()
```

```python
# BAR PLOT

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(d['day'],d['tip'], data = d)

```

```python
# HISTOGRAM

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.hist(d['tip'])

ax.grid(True)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.hist(d['tip'])

ax.grid(True)
ax.set_xticks([1,3,5,7,9])
```

```python
# BOXPLOT
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.boxplot('total_bill'  ,data =d)

ax.grid(True)
```

```python
# SCATTERPLOT

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.scatter(d['tip'],d['total_bill'])

ax.grid(True)
```

```python
fig, axis = plt.subplots(1,4,figsize = (12,4))

axis[0].bar(d['day'],d['tip'])

axis[1].hist(d['tip'])

axis[2].scatter(d['tip'],d['total_bill'])

axis[3].boxplot(d['tip'])
```

```python
fig, axis = plt.subplots(1,4,figsize = (19,6), dpi =150)

axis[0].bar(d['day'],d['tip'])
axis[0].grid(True)
axis[0].set_title("Barplot")
axis[0].set_xlabel("day")
axis[0].set_ylabel("tip")

axis[1].hist(d['tip'])
axis[1].grid(True)
axis[1].set_title("Histogram")
axis[1].set_xlabel("tip")
axis[1].set_ylabel("freq")

axis[2].scatter(d['tip'],d['total_bill'])
axis[2].grid(True)
axis[2].set_title("Scatterplot")
axis[2].set_xlabel("tip")
axis[2].set_ylabel("total_bill")

axis[3].boxplot(d['tip'])
axis[3].grid(True)
axis[3].set_title("Boxplot")
axis[3].set_xlabel("tip")
axis[3].set_ylabel("?")
```

```python
fig, axis = plt.subplots(2,2,figsize = (19,15), dpi =150)

axis[0][0].bar(d['day'],d['tip'])
axis[0][0].grid(True)
axis[0][0].set_title("Barplot")
axis[0][0].set_xlabel("day")
axis[0][0].set_ylabel("tip")

axis[0][1].hist(d['tip'])
axis[0][1].grid(True)
axis[0][1].set_title("Histogram")
axis[0][1].set_xlabel("tip")
axis[0][1].set_ylabel("freq")

axis[1][0].scatter(d['tip'],d['total_bill'])
axis[1][0].grid(True)
axis[1][0].set_title("Scatterplot")
axis[1][0].set_xlabel("tip")
axis[1][0].set_ylabel("total_bill")

axis[1][1].boxplot(d['tip'])
axis[1][1].grid(True)
axis[1][1].set_title("Boxplot")
axis[1][1].set_xlabel("tip")
axis[1][1].set_ylabel("?")
```
